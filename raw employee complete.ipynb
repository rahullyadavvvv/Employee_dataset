```python
import pandas as pd
df_raw = pd.read_csv('raw_employee_data_one_row.csv' , header= None)
print("raw shape:" , df_raw.shape)
df_raw.head()
```

```python

columns = ['Name', 'Roll' , 'Age' , 'Gender' , 'Field' , 'Salary' , 'Address']
data_list = df_raw.values.flatten().tolist()
records_1 = [data_list[i:i+7] for i in range(0 , len(data_list),7)]
print(records_1)
df_clean_1 = pd.DataFrame(records_1 , columns=columns)
print("clean shape:" , df_clean_1.shape)
df_clean_1.head()
df_clean_1.to_csv('data_clean_1.csv' , index = False)
print(df_clean_1)
df_clean_1.head()

```

```python
df_male = df_clean_1[df_clean_1['Gender'] == 'Male']
df_female = df_clean_1[df_clean_1['Gender'] == 'Female']
df_other = df_clean_1[df_clean_1['Gender'] == 'Other']

print("Male students:")
print(df_male)

print("\nFemale students:")
print(df_female)

print("\nOther students:")
print(df_other)
```

```python
average = df_clean_1['Salary'].mean()
print(average)
```

```python
import matplotlib.pyplot as plt

df_clean_1['Salary'] = pd.to_numeric(df_clean_1['Salary'], errors='coerce')

avg_salary = df_clean_1.groupby('Roll')['Salary'].mean()

avg_salary.plot(kind='bar', color='green')
plt.title('Average Salary by Skills Role')
plt.ylabel('Average Salary')
plt.xlabel('Job Role')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
gender_counts = df_clean_1['Gender'].value_counts()

gender_counts.plot(kind='pie', autopct='%1.1f%%', colors=['lightpink', 'lightblue', 'lavender'], startangle=140)
plt.title('gender' )
plt.ylabel('')
plt.show()
```
